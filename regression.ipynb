# SIMPLE LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
dataset = pd.read_csv('/kaggle/input/salary-data-for-simple-linear-regression/Salary_Data.csv')
dataset
```

```python
#Getting the independent and the dependent variables
#X -> Years of experience
X = dataset.iloc[: ,:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#Splitting the data into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)
```

```python
#Training the simple linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predict the test
y_pred = regressor.predict(X_test)
```

```python
#Visualization
plt.subplot(1,2,1)
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title("Salary vs Experience (Training Set)")

plt.subplot(1,2,2)
plt.scatter(X_test, y_test, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title("Salary vs Experience (Test Set)")


plt.tight_layout()
plt.show()
```

# MULTIPLE LINEAR REGRESSION

```python
#Import the datasets
dataset = pd.read_csv('/kaggle/input/data-of-50-startups-for-multiple-linear-regression/50_Startups.csv')
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#Encoding categorical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder="passthrough")
X = np.array(ct.fit_transform(X))
X
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
#Modelling
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predictions 
y_pred = regressor.predict(X_test)
error = y_test - y_pred
np.set_printoptions(precision=2)
print(np.concatenate((y_test.reshape(len(y_pred),1), y_pred.reshape(len(y_test),1), error.reshape(len(y_pred),1)), axis=1))
```

```python
#Visualization
plt.subplot(1,3,1)
plt.scatter(X_test[:, 3], y_test, color="red")
plt.plot(X_test[:,3], y_pred, color="blue")

plt.subplot(1,3,2)
plt.scatter(X_test[:, 4], y_test, color="red")
plt.plot(X_test[:,4], y_pred, color="blue")

plt.subplot(1,3,3)
plt.scatter(X_test[:, 5], y_test, color="red")
plt.plot(X_test[:,5], y_pred, color="blue")

plt.tight_layout()
plt.show()
```

# Polynomial Linear Regression

```python
dataset = pd.read_csv('/kaggle/input/polynomial-regression-data/Position_Salaries.csv')
dataset
```

```python
#X and y
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
#Linear Regression model for comparing
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
#Create the features 
from sklearn.preprocessing import PolynomialFeatures
poly_feature = PolynomialFeatures(degree=4)
X_poly = poly_feature.fit_transform(X)
```

```python
X_poly
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
#Vizualization
plt.scatter(X, y, color="pink")
plt.plot(X, lin_reg.predict(X), color="red")
plt.plot(X, poly_reg.predict(X_poly), color="cyan")
plt.title("Checking the Salary for Position Level")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

```python
#Predicting values
lin_reg.predict([[6.5]])
```

```python
#Predicting value with Polynomial 
X_poly_test = poly_feature.fit_transform([[6.5]])
poly_reg.predict(X_poly_test)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
X
```

# Decision Tree Regression

```python
dataset = pd.read_csv('/kaggle/input/polynomial-regression-data/Position_Salaries.csv')
dataset
```

```python
X = dataset.iloc[:, 1:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
#Training the model
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

```python
#Predicting new value
regressor.predict([[6.5]])
```

```python
#Visualization 
plt.scatter(X,y, color="yellow")
plt.plot(X, regressor.predict(X), color="pink")
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X,y)
```

```python
#Predict
regressor.predict([[6.5]])
```

```python
#Visualization 
plt.scatter(X,y, color="yellow")
plt.plot(X, regressor.predict(X), color="pink")
```

# Evaluation

```python
from sklearn.metrics import r2_score
r2_score(y, regressor.predict(X))
```
